```python
def sample_mlp(rs, in_dim, out_dim, hidden_dims, nonlinearity):
    """Sample an MLP with given dimensions and nonlinearity.

    Args:
        rs: Random state for reproducibility.
        in_dim: Input dimension.
        out_dim: Output dimension.
        hidden_dims: List of hidden layer dimensions.
        nonlinearity: Nonlinearity function to use between layers.

    Returns:
        A function representing the MLP.
    """
    layers = []
    layer_dims = [in_dim] + hidden_dims + [out_dim]
    
    for i in range(len(layer_dims) - 1):
        W = rs.normal(size=(layer_dims[i], layer_dims[i+1])) * np.sqrt(2. / layer_dims[i])
        b = rs.normal(size=(layer_dims[i+1],)) * 0.01
        layers.append((W, b))
    
    def mlp(x):
        for i, (W, b) in enumerate(layers):
            x = x @ W + b
            if i < len(layers) - 1:
                x = nonlinearity(x)
        return x
    
    return mlp
```

```python
def sample_matrix_product(rs, dim, n, distribution='normal'):
    """Sample a product of n random matrices of size (dim, dim).

    Args:
        rs: Random state for reproducibility.
        dim: Dimension of the square matrices.
        n: Number of matrices to multiply.
        distribution: Distribution to sample from ('normal' or 'uniform').

    Returns:
        The resulting matrix product.
    """
    if distribution == 'normal':
        matrices = [rs.normal(size=(dim, dim)) for _ in range(n)]
    elif distribution == 'uniform':
        matrices = [rs.uniform(size=(dim, dim)) for _ in range(n)]
    else:
        raise ValueError("Unsupported distribution type.")

    M_prod = np.eye(dim)
    for M in matrices:
        M_prod = M_prod @ M
    return M_prod
```
