```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error

from cbm.SCBM_models import get_SCBM_tf_1
from cbm.estimation import estimate_bottleneck_and_mechanism_fcts
```

```python
# GLOBAL PARAMS
seed = 0
n_samples = 100
d_micro = 80
```

```python
test_scbm = get_SCBM_tf_1(seed=seed, d=d_micro)
train_scbm = get_SCBM_tf_1(seed=seed, d=d_micro)
```

```python
test_samples, test_bn_samples = test_scbm.sample(size=1000)
train_samples, train_bn_samples = test_scbm.sample(size=n_samples)

bn_train_samples, bn_train_bn_samples = train_scbm.sample(size=100)
```

```python
# OLS w/ conditioning on node
regr_node = LinearRegression()

regr_node.fit(np.concatenate([train_samples[1], train_samples[0]], axis=1), train_samples[2])

y_hat_node = regr_node.predict(np.concatenate([test_samples[1], test_samples[0]], axis=1))
```

```python
# OLS w/ conditioning on g.t. bottleneck
regr_bn = LinearRegression()

regr_bn.fit(np.concatenate([train_samples[1], train_bn_samples[0, 1]], axis=1), train_samples[2])

y_hat_bn = regr_bn.predict(np.concatenate([test_samples[1], test_bn_samples[0, 1]], axis=1))
```

```python
# OLS w/ conditioning on estimated bottleneck (from other environment)
estimated_bn_fcts , _ = estimate_bottleneck_and_mechanism_fcts(train_scbm, bn_train_samples, mode='linear')

regr_bn = LinearRegression()

estimated_bn_train = estimated_bn_fcts[0, 1](train_samples[0])
regr_bn.fit(np.concatenate([train_samples[1], estimated_bn_train], axis=1), train_samples[2])

estimated_bn_test = estimated_bn_fcts[0, 1](test_samples[0])
y_hat_bn = regr_bn.predict(np.concatenate([test_samples[1], estimated_bn_test], axis=1))
```

```python
# Get metrics
y = test_samples[2]

# Node conditioning
mse_node = mean_squared_error(y, y_hat_node)
mae_node = mean_absolute_error(y, y_hat_node)

error_node = y - y_hat_node
var_node = np.var(error_node)

# Bottleneck conditioning
mse_bn = mean_squared_error(y, y_hat_bn)
mae_bn = mean_absolute_error(y, y_hat_bn)

error_bn = y - y_hat_bn
var_bn = np.var(error_bn)
```

```python
print('Node:')
print(f'MSE: {mse_node}')
print(f'MAE: {mae_node}')
print(f'Var: {var_node}')

print('Bottleneck:')
print(f'MSE: {mse_bn}')
print(f'MAE: {mae_bn}')
print(f'Var: {var_bn}')
```
